#### Creating visualizations of queried shipment info using Seaborn

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
# use this to customize the y-axis label
from matplotlib.ticker import FuncFormatter
```

##### Data file paths

```python
# US port name, Total Weight (kg), Total Volume (m3), Number of Shipments
US_port_file_path = r"view_query_data\total_shipment_us_ports_2018_to_2020.csv"

# Foreign port name, Total Weight (kg), Total Volume (m3), Number of Shipments
# Foreign_port_file_path = r"notebooks_visualizations\view_query_data\shipment_info_top_foreign_ports.csv"

# Year, Month, 
Monthly_file_path = r"view_query_data\total_shipment_year_month.csv"
```

##### Line plot for total shipments per month through 2018-2020

```python
monthly_df = pd.read_csv(Monthly_file_path)
sns.lineplot(
    data=monthly_df, 
    x=monthly_df['arrival_month'].astype(str) + '-' + monthly_df['arrival_year'].astype(str),
    y='shipment_count'
)
plt.xlabel("Month-Year")
plt.ylabel("Number of Shipments (Millions)")
plt.title("Monthly Shipments to US Ports (2018-2020)")
plt.xticks(rotation=80)
plt.title("Monthly Shipments to US Ports (2018-2020)")
plt.show()
```

```python
# Read CSV (no header)
df = pd.read_csv(US_port_file_path, names=['us_port','shipment_count'], header=0)

sns.barplot(data=df, x='us_port', y='shipment_count')
# rotate port name
plt.xticks(rotation=45, ha='right')

plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{int(x)/1000000}m'))

plt.title('Top 15 US Ports (2018-2020)')
plt.xlabel('Category')
plt.ylabel('Shipment Count (Millions)')
plt.show()
```

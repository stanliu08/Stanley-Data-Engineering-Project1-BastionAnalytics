##### Concatenating

```python
# combined header file
output = r"..\data\bronze_raw\CBP AMS Shipping Data\raw-layer\ams_header_combined.csv"
chunk_size = 500000
```

##### Time took for concatenating the files: 9m 49.1s

```python
# new updated chunking for all years
first = True

for year, file_list in files.items():
    for f in file_list:
        for chunk in pd.read_csv(f, chunksize=chunk_size):
            
            # add the year column
            chunk["source_year"] = year

            # write chunk to output
            chunk.to_csv(
                output,
                mode="a",
                index=False,
                header=first
            )
            first = False
```
